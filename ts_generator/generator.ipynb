```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
date_range = pd.date_range(start='1/1/2023', end='31/12/2024', freq='D')
```

```python
macro_trend = np.linspace(50, 100, len(date_range))  # Example: linear trend
```

```python
shop_trend = np.log(np.arange(1, len(date_range) + 1)) * 5  # Example: logarithmic trend
```

```python
yearly_seasonality = 20 * np.sin(np.linspace(0, 2 * np.pi * 3, len(date_range)))  # 3 cycles over 3 years
```

```python
weekly_seasonality = 10 * np.sin(np.linspace(0, 2 * np.pi * (len(date_range) / 7), len(date_range)))  # weekly cycle
```

```python
noise = np.random.normal(loc=0, scale=3, size=len(date_range))
```

```python
sales = macro_trend + shop_trend + yearly_seasonality + weekly_seasonality + noise
```

```python
time_series_data = pd.DataFrame({'Date': date_range, 'Sales': sales})
time_series_data.set_index('Date', inplace=True)

time_series_data.plot(figsize=(15, 8))
plt.title('Synthetic Time Series Data with Multiple Components')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.show()
```

```python
plt.figure(figsize=(15, 12))

# Macro Economic Trend
plt.subplot(5, 1, 1)
plt.plot(date_range, macro_trend, label='Macro Economic Trend')
plt.legend()

# Shop-Specific Trend
plt.subplot(5, 1, 2)
plt.plot(date_range, shop_trend, label='Shop-Specific Trend')
plt.legend()

# Yearly Seasonality
plt.subplot(5, 1, 3)
plt.plot(date_range, yearly_seasonality, label='Yearly Seasonality')
plt.legend()

# Weekly Seasonality
plt.subplot(5, 1, 4)
plt.plot(date_range, weekly_seasonality, label='Weekly Seasonality')
plt.legend()

# Noise
plt.subplot(5, 1, 5)
plt.plot(date_range, noise, label='Noise')
plt.legend()

plt.tight_layout()
plt.show()
```
