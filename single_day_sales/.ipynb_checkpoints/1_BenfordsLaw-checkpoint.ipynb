```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_data_types = {
    '会员卡号': str,  # Replace 'Column1' with the actual name of the first column
    '外卖订单号': str   # Replace 'Column2' with the actual name of the second column
}
df = pd.read_excel('dataset/majiyong.xlsx', sheet_name='Sheet1', dtype=column_data_types)
df.dataframeName = 'majiyong'
df['Profit'] = df['实收金额'] - df['成本金额']
```

```python
df.head(10)
```

```python
df = df[df['实收金额'] != 0]
```

```python
def extract_leading_digit(num):
    num = abs(num)  # Handle negative numbers
    while num < 1 and num != 0:
        num *= 10
    return int(str(num)[0])

df['leading_digit'] = df['实收金额'].apply(extract_leading_digit)
```

```python
df['leading_digit']
```

```python
all_digits = pd.Series(index=np.arange(1, 10), dtype=float)
observed_frequencies = df['leading_digit'].value_counts(normalize=True)
observed_frequencies = all_digits.combine_first(observed_frequencies).fillna(0)

# Calculate expected frequencies using Benford's Law
digits = np.arange(1, 10)
expected_frequencies = np.log10(1 + 1/digits)

# Plot observed vs expected frequencies
plt.figure(figsize=(10, 6))
plt.bar(digits - 0.15, observed_frequencies, width=0.3, label='Observed')
plt.bar(digits + 0.15, expected_frequencies, width=0.3, label='Benford')
plt.xticks(digits)
plt.xlabel('Leading Digit')
plt.ylabel('Frequency')
plt.legend()
plt.title('Benford’s Law Analysis')
plt.show()
```

```python
from scipy.stats import chisquare
chi_statistic, p_value = chisquare(observed_frequencies, f_exp=expected_frequencies)
print(chi_statistic, p_value)
```

```python
# Chi-square Statistic: 0.38552505885238225
# P-value: 0.9999506685129054
# Interpreting these results:

# Chi-square Statistic: The value is relatively low, suggesting that the observed frequencies of leading digits in your dataset are not significantly different from what would be expected based on Benford's Law.

# P-value: A very high p-value (close to 1) strongly indicates that there is no significant difference between your observed data and the expected frequencies as per Benford's Law. In other words, your data closely follows the distribution predicted by Benford's Law.
```

```python
MAD = np.mean(np.abs(observed_frequencies - expected_frequencies))
print(MAD)
```

```python
# The Mean Absolute Deviation (MAD) value of approximately 0.053 indicates that, on average, the observed frequencies of the leading digits in your data deviate 
# from the expected frequencies as per Benford's Law by about 5.3%. 
# This suggests a relatively small deviation, which could be considered as an indication of natural variance in data that generally conforms to Benford's Law.
```

```python
from scipy.stats import ks_2samp
D_statistic, p_value = ks_2samp(observed_frequencies.cumsum(), expected_frequencies.cumsum())
print(D_statistic, p_value)
```

```python
# D-Statistic (0.222): This value represents the maximum distance between the cumulative distribution functions (CDFs) of the observed and expected frequencies. A value of 0.222 indicates that the maximum deviation between these two CDFs is about 22.2%.

# P-Value (0.989): This is a very high p-value, which indicates that the differences between the observed and expected distributions are not statistically significant. In statistical terms, we fail to reject the null hypothesis that the two distributions are the same.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate 10000 random numbers
random_numbers = np.random.uniform(low=1, high=10000, size=10000)

# Function to extract the leading digit
def leading_digit(num):
    while num >= 10 or num <= -10:
        num /= 10
    return int(abs(num))

# Extract leading digits
leading_digits = [leading_digit(num) for num in random_numbers]

# Convert to DataFrame for analysis
df = pd.DataFrame({'leading_digit': leading_digits})

# Calculate observed frequencies
all_digits = pd.Series(index=np.arange(1, 10), dtype=float)
observed_frequencies = df['leading_digit'].value_counts(normalize=True)
observed_frequencies = all_digits.combine_first(observed_frequencies).fillna(0)

# Calculate expected frequencies using Benford's Law
digits = np.arange(1, 10)
expected_frequencies = np.log10(1 + 1/digits)

# Plot observed vs expected frequencies
plt.figure(figsize=(10, 6))
plt.bar(digits - 0.15, observed_frequencies, width=0.3, label='Observed')
plt.bar(digits + 0.15, expected_frequencies, width=0.3, label='Benford')
plt.xticks(digits)
plt.xlabel('Leading Digit')
plt.ylabel('Frequency')
plt.legend()
plt.title('Benford’s Law Analysis on Random Numbers')
plt.show()
```
